- ## Example combining with reduce()
 - 2 inputs to 1 output, which is repeated
 - rule for 2 inputs to 1 output, is given by 1st parameter of the function

```python
import functools

a = [1,2,3,4,5]
b = functools.reduce(lambda x,y:x+y, a)
print(b)

#[process details]
#(1) 1 and 2 are summed by the lambda function -> [3,3,4,5]
#(2) 3 and 3 become 6 -> [6,4,5]
#(3) 6 and 4 become 10 -> [10,5]
#(4) 10 and 5 become 15
```

## Practice

- ## average : mean()
- ## prime number : is_prime()
- ## number of prime numbers : num_prime()

```python
def mean(nums):
    _sum = 0
    for i in nums:
        _sum += i
    return (_sum/len(nums))

a = range(10)
print(list(a))
print(tuple(a))
b = mean(a)
print(b)
```

```python
def is_prime(num):
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

print(is_prime(5))
print(is_prime(25))
print(is_prime(37))
```

```python
def num_prime(num):
    count = 0
    for i in range(2, num+1):
        if is_prime(i):
            count += 1
    return count

print(num_prime(5))
print(num_prime(7))
print(num_prime(100))
```

- ## Members and their initialization

```python
class Person:
    def __init__(self, name, age = 10):  # members :name, age, hobby
        print(self, "is generated")      # name : updated by the parameter 'name'
        self.name = name                 # age : updated by the parameter 'age', and
        self.age = age                   # if no parameter 'age', it becomes the default value '10'
        self.hobby = 'singing'           # hobby : set to 'singing'
```

- ## Instance method

```python
class Counter:
    def __init__(self, num):
        self.num = num
    def Increment(self):
        self.num += 1
    def Reset(self):
        self.num = 0
    def InputValue(self, num): # similar to __init(), 'self' is a must and followed by parameters
        self.num = num
    def PrintCurrentValue(self):
        print('value = ',self.num)
        
```

```python
c1 = Counter(0)   # creation of 2 objects c1, c2
c2 = Counter(10)
c1.PrintCurrentValue()
c2.PrintCurrentValue()
```

```python
c1.Increment()
c1.Increment()
c2.Increment()
```

```python
c1.PrintCurrentValue()
c2.PrintCurrentValue()
```

```python
class Computer:
    def add(self, x, y):
        return x+y
    def sub(self,x,y):
        return x-y

# In this case, the object 'c1' was created just to do add() and sub()
# -> with no object, we want to do add() adn sub()
```

```python
class Computer:
    @staticmethod
    def add(x,y):
        return x+y
    @staticmethod
    def sub(x,y):
        return x-y
```

```python
print(Computer.add(10,20))
print(Computer.sub(10,20))
```

- ## Nested class (class as a class's member)

```python
class Dept:
    def __init__(self, dname):
        self.dname = dname
    class Prof:
        def __init__(self, pname):
            self.pname = pname
math = Dept("Mathematics")
mathprof = Dept.Prof("Mark")

print(math.dname)
print(mathprof.pname)
```

```python
class Dept:
    def __init__(self, dname):
        self.dname = dname
        self.inner = self.Prof()
    def outer_disp(self):
        self.inner.inner_disp(self.dname)
    class Prof:
        def inner_disp(Self, details):
            print(details, "From Inner Class")
            
math = Dept("Mathmatics")

math.outer_disp()
```

- ## Method Overriding
 - In case where a method in child class is the same in parent class, the method in child class is activated
 - The overridden method of parent class cannot be used by child class

```python
class Student(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def Work(self, hour):
        print('{}는 {}시간동안 공부한다.'.format(self.name, hour))

class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
bob = Student('Bob', 30)  # child class 'Student' does not have Eat(), Sleep(), Work(),
bob.Eat('BBQ')            # but inherits them from the parent class 'Person'
bob.Sleep(8)
bob.Work(10)
```

- ## The overridden method of parent class is activated by 'super()'

```python
class Student(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def Work(self, hour):
        super().Work(hour)
        print('{}는 {}시간동안 공부한다.'.format(self.name, hour))

class Employee(Person):
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
bob = Student('Bob', 30)  # child class 'Student' does not have Eat(), Sleep(), Work(),
bob.Eat('BBQ')            # but inherits them from the parent class 'Person'
bob.Sleep(8)
bob.Work(10)
```

- ## Redefinition of addition operator : __add__()

```python
class Point2D:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def Add(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point2D(new_x, new_y)
    def __str__(self):
        return ('({}, {})'.format(self.x, self.y))

p1 = Point2D(3,5)
p2 = Point2D(1,-1)

p3 = p1.Add(p2)

print(p1)
print(p2)
print(p3)
```

```python
class Point2D:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point2D(new_x, new_y)
    def __str__(self):
        return ('({}, {})'.format(self.x, self.y))

p1 = Point2D(3,5)
p2 = Point2D(1,-1)

p3 = p1 + p2

print(p1)
print(p2)
print(p3)
```

- ## Redefinition of subtraction operator : __sub__()

```python
class Point2D:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self, pt):
        new_x = self.x + pt.x
        new_y = self.y + pt.y
        return Point2D(new_x, new_y)
    def __sub__(self, pt):
        new_x = self.x - pt.x
        new_y = self.y - pt.y
        return Point2D(new_x, new_y)
    def __str__(self):
        return ('({}, {})'.format(self.x, self.y))

p1 = Point2D(3, 5)
p2 = Point2D(1, -1)

p3 = p1 + p2
p4 = p1 - p2 

print(p1)
print(p2)
print(p3)
print(p4)
```

- ## Class for complex number system

```python
class ComplexNumber:
    def __init__(self, real, img):
        self.real = real
        self.img = img
    def __str__(self):
        if self.img>=0:
            return '{}+j{}'.format(self.real, self.img)
        else:
            return '{}-j{}'.format(self.real, abs(self.img))
    def __add__(self,cn):
        return ComplexNumber(self.real+cn.real, self.img+cn.img)
    def __sub__(self,cn):
        return ComplexNumber(self.real-cn.real, self.img-cn.img)
```

```python
a = ComplexNumber(1,2)
b = ComplexNumber(-3,3)
c = a + b
d = a - b
print(a,b,c,d, sep=', ')
```
