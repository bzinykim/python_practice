# 과제 : Class Practice

```python
class ComplexNumber:
    def __init__(self, real, img):
        self.real = real
        self.img = img
    def cprint(self):
        if self.img >= 0:
            return '{}+j{}'.format(self.real, self.img)
        else:
            return '{}-j{}'.format(self.real, abs(self.img))
    def cadd(self,cn):
        return ComplexNumber(self.real+cn.real, self.img+cn.img)
    def csub(self,cn):
        return ComplexNumber(self.real-cn.real, self.img-cn.img)
```

```python
a = ComplexNumber(1,2)
b = ComplexNumber(-3,3)
c = a.cadd(b)# c = a + b
d = a.csub(b)# d = a -  b
a.cprint(),b.cprint(),c.cprint(),d.cprint() # print(a, b, c, d, esp=', ')
```

### Graph overlapping
 - multiple use of 'plot()' leads to graph overlapping
 - To give a name to a curve, use 'label' in 'plot()' and execute 'lengend()'

```python
data = np.random.randn(30)
plt.plot(data, 'k--', label='default')
plt.plot(data, 'k-', drawstyle='steps-post', label ='steps=pos')
plt.legend()
plt.show()
```

### Graph title, axis label

```python
data = np.random.randn(100)
plt.title('random')
plt.xlabel('index')
plt.ylabel('height')
plt.plot(data, 'rx--')
plt.show()
```
